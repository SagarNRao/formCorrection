```python
# --- 1_Laptop_RTSP_Server.ipynb ---
# Run this on your Windows laptop

import subprocess
import time
import os
from IPython.display import display, HTML

# ------------------ CONFIG ------------------
WEBCAM_NAME = "Lenovo FHD Webcam"           # Change: run `ffmpeg -list_devices true -f dshow -i dummy` to find yours
LAPTOP_IP = "10.227.207.170"        # Your laptop's IP (must be reachable from RPi)
RTSP_PORT = 8554
STREAM_PATH = "webcam.sdp"
# -------------------------------------------

RTSP_URL = f"rtsp://{LAPTOP_IP}:{RTSP_PORT}/{STREAM_PATH}"
print(f"Stream will be available at: {RTSP_URL}")

# Build FFmpeg command
ffmpeg_cmd = [
    "ffmpeg",
    "-f", "dshow",
    "-i", f"video={WEBCAM_NAME}",
    "-c:v", "libx264",
    "-preset", "ultrafast",
    "-tune", "zerolatency",
    "-profile:v", "baseline",
    "-pix_fmt", "yuv420p",
    "-b:v", "1000k",
    "-bufsize", "2000k",
    "-g", "30",
    "-f", "rtsp",
    "-rtsp_transport", "tcp",
    RTSP_URL
]

print("\nStarting RTSP server... (Press Ctrl+C to stop)\n")
print(" ".join(ffmpeg_cmd))
print("\n" + "="*60)

# Start FFmpeg as subprocess
process = subprocess.Popen(
    ffmpeg_cmd,
    stdout=subprocess.PIPE,
    stderr=subprocess.STDOUT,
    text=True,
    bufsize=1
)

# Live log output
try:
    for line in process.stdout:
        print(line.strip())
except KeyboardInterrupt:
    print("\nStopping RTSP server...")
    process.terminate()
    process.wait()
    print("Server stopped.")
```
